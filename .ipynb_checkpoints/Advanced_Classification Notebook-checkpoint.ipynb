### 3.4.3 The Influence of Twitter Handles on Sentiment Classes

The username at the end of your unique Twitter URL is known as a Twitter handle.
A Twitter handle and a Twitter name are not the same thing. After the @ sign in your profile URL, you'll see your Twitter handle, which should be unique to your account. A Twitter handle, on the other hand, exists solely to assist people in finding the company/individualsthey seek (Elizabeth Arens, 2019).

```python
# Creating a function to extract handles from tweets
def extract_handles(x):
    """ This function finds handles in a 
        tweet and returns them as a list"""
    handles = []
    for i in x:
        h = re.findall(r'@(\w+)', i)
        handles.append(h)
        
    return handles
```

```python
# Extracting handles from tweets
news_h = extract_handles(tweets_df['message']
                              [tweets_df['sentiment'] == 2])
pro_h = extract_handles(tweets_df['message']
                          [tweets_df['sentiment'] == 1])
neutral_h = extract_handles(tweets_df['message']
                              [tweets_df['sentiment'] == 0])
anti_h = extract_handles(tweets_df['message']
                          [tweets_df['sentiment'] == -1])

# Unnesting list
handles = [sum(news_h, []), sum(pro_h, []), sum(neutral_h, []),
           sum(anti_h, [])]

# Visualising the Handles
full_title = ['Impact of Handles on the News sentiment',
              'Impact of Handles on the Pro sentiment',
              'Impact of Handles on the Neutral sentiment',
              'Impact of Handles on the Anti sentiment']

plt.rcParams['figure.figsize'] = [50, 5]

for i, sent in enumerate(handles):
    plt.subplot(1, 4, i + 1)
    freq_dist = nltk.FreqDist(sent)
    df = pd.DataFrame({'Handle': list(freq_dist.keys()),
                      'Count' : list(freq_dist.values())})

    df = df.nlargest(columns='Count', n=10)

    ax = sns.barplot(data=df, y='Handle', x='Count', palette="tab20c")
    plt.title(full_title[i])
    plt.show()
```

### 3.4.4 The Influence of Twitter Hashtags on Sentiment Classes

A hashtag is a word or phrase that begins with the hash symbol (#). It's used within a social media post to help people who might be interested in your topic find it when they search for a keyword or a hashtag. It aids in attracting attention to your posts and encouraging participation (Clodagh OBrien, 2022).

```python
# Creating a function to extract hashtags from tweets
def extract_hashtags(x):
    """ The following function finds hashtags on a 
        tweet and returns them as a list"""
    hashtags = []
    for i in x:
        ht = re.findall(r'#(\w+)', i)
        hashtags.append(ht)
        
    return hashtags
```

## 4.1 Cleaning the raw tweets dataframe with the Regex library

Regular expressions also known as REs, regexes, or regex patternsare a small, highly specialized programming language that is embedded in Python and accessible through the re module. You specify the rules for the set of possible strings that you want to match using this little language; this set could include English sentences, e-mail addresses, TeX commands, or anything else you want. REs can also be used to modify or split a string in a variety of ways (A.M. Kuchling, 2022).

### 4.1.1 Noise Removal

The most important part of getting data into usable format in text analysis is removing noise.
Noise in data is any text that is unrelated to the context of the data.. URLs and links, social media entities (mentions, hashtags), Language Stopwords, and punctuation can all be identified as noise. 


The Function below will help us get rid of the noise.
- Convert letters to lowercases
- remove URL links
- remove # from hashtags
- remove numbers
- remove punctuation

```python
def cleaner(tweet):
    """
    this function takes in a dataframe and perform the following:
    -Convert letters to lowercases
    -remove URL links
    -remove # from hashtags
    -remove numbers
    -remove punctuation
    from the text field then return a clean dataframe 
    """
    tweet = tweet.lower() #convert to text to lowercase
    to_del = [
        r"@[\w]*",  # strip account mentions
        r"http(s?):\/\/.*\/\w*",  # strip URLs
        r"#\w*",  # strip hashtags
        r"\d+",  # delete numeric values
        r"U+FFFD",  # remove the "character note present" diamond
    ]
    for key in to_del:
        tweet = re.sub(key, "", tweet)
    
    # strip punctuation and special characters
    tweet = re.sub(r"[,.;':@#?!\&/$]+\ *", " ", tweet)
    
    # strip excess white-space
    tweet = re.sub(r"\s\s+", " ", tweet)
    tweet = re.sub(r'rt[\s]+', '', tweet) #Remove RT
    
    return tweet.lstrip(" ")
```

```python
train_df['message'] = train_df['message'].apply(cleaner)  #apply function to remove noise in the data
test_df['message'] = test_df['message'].apply(cleaner)  #apply function to remove noise in the data
```

## 4.2 Tokenisation

Tokenization is the process of dividing a piece of text into smaller units known as tokens. Tokens can be words, characters, or subwords in this context. Each word is divided into a token using the tokenizer.

```python
def tokenize(txt): #function to tokenize  our dataset
    """
    Splits a words into smaller units.
    
    Parameyters:
    txt: a string(a sequence of characters)
    
    returns:
    tokens: componets of a text
    """
    tokens = re.split('\W+', txt)
    return tokens
train_df['tokens'] = train_df['message'].apply(tokenize)
test_df['tokens'] = test_df['message'].apply(tokenize)
```

## 4.3 Removing Stop words

A stop word is a commonly used word (such as "the," "a," "an," or "in") that a search engine has been programmed to ignore, both when indexing and retrieving entries as the result of a search query.
We don't want these words to take up valuable storage space or processing time in our database. We can easily remove them by keeping a list of words you consider to be stop words.

```python
# sample stop words
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
def remove_stopwords(txt_tokens):
    """
    Removes stop words from tokens
    
    Parameters:
    txt_tokens: components of a text
    
    Returns:
    txt_clean: a list containing tokens without stopwords
    """
    txt_clean = [word for word in txt_tokens if word not in stopwords]
    return txt_clean

train_df['no_stopwords'] = train_df['tokens'].apply(remove_stopwords)
test_df['no_stopwords'] = test_df['tokens'].apply(remove_stopwords)
```

```python
train_df.tail()
```

## 4.4 Stemming

Stemming algorithms function by removing the beginning or end of the word while taking into account a list of frequent prefixes and suffixes that can be found in an inflected word.Stemming is simpler, but it frequently fails to yield a morphological root of the word that is unique to a dictionary.

```python
def stemming(no_stopwords):
    """
    Removes suffixes of words
    
    Parameters:
    no_stopwords: a list of words without stopwords
    
    Returns:
    text: a list of words that have been reduced
    """
    text = [ps.stem(word) for word in no_stopwords]
    return text

train_df['stems'] = train_df['no_stopwords'].apply(stemming) #adding a stems column to the train_df dataframe
test_df['stems'] = test_df['no_stopwords'].apply(stemming) #adding a stems column to the test_df dataframe
```

```python
train_df.head()
```

## 4.5 Lemmatisation

On the other hand, lemmatization takes into account the morphological analysis of the words. To do this, it is essential to have comprehensive dictionaries that the algorithm can use to connect the form to its lemma. With the same train data, we can see how it functions.

```python
def lemmatization(no_stopwords):
    """
    Reduces words to their root form
    
    Parameters:
    no_stopwords: a list of words without stopwords
    
    Returns:
    text: a list with words which have been reduces to their root form
    """
    text = [wn.lemmatize(word) for word in no_stopwords]
    return text

train_df['lemmas'] = train_df['no_stopwords'].apply(lemmatization) #Adding the column lemmas to the df_train dataframe
test_df['lemmas'] = test_df['no_stopwords'].apply(lemmatization) #Adding the column lemmas to the df_test dataframe
```

```python
def form_sentence(tweet):
    tweet_blob = TextBlob(tweet)
    return ' '.join(tweet_blob.words)

def no_user_alpha(tweet):
    tweet_list = [ele for ele in tweet.split() if ele != 'user']
    clean_tokens = [t for t in tweet_list if re.match(r'[^\W\d]*$', t)]
    clean_s = ' '.join(clean_tokens)
    clean_mess = [word for word in clean_s.split() if word.lower() not in stopwords.words('english')]
    return clean_mess

def normalization(tweet_list):
        lem = WordNetLemmatizer()
        normalized_tweet = []
        for word in tweet_list:
            normalized_text = lem.lemmatize(word,'v')
            normalized_tweet.append(normalized_text)
        return normalized_tweet

def wrangle(df,post_column):
    df['tweet_list'] = df[post_column].apply(no_user_alpha(form_sentence()))
    df['normalized_tweet'] = df['tweet_list'].apply(normalization())

    return df
```

```python
names = ['Logistic Regression',
          'RBF SVM',          
         'Decision Tree', 'Random Forest',  'AdaBoost','RidgeClassifier']
```

```python
classifiers = [
    LogisticRegression(), 
    SVC(C= 3.43714213619288, degree= 1, gamma= 0.294916193137584, kernel= 'rbf'),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),    
    AdaBoostClassifier(),
    RidgeClassifier(),
    RidgeClassifierCV()
]
```

```python
results = []

models = {}
confusion = {}
class_report = {}


for name, clf in zip(names, classifiers):    
    print ('Fitting {:s} model...'.format(name))
    run_time = %timeit -q -o clf.fit(X_train, y_train)
    
    print ('... predicting')
    y_pred = clf.predict(X_train)   
    y_pred_test = clf.predict(X_test)
    
    print ('... scoring')
    accuracy  = accuracy_score(y_train, y_pred)
    precision = precision_score(y_train, y_pred,pos_label='positive'
                                           ,average='micro')
    recall    = recall_score(y_train, y_pred,pos_label='positive'
                                           ,average='micro')

    
    f1        = f1_score(y_train, y_pred,average='micro')    
    f1_test   = f1_score(y_test, y_pred_test,average='micro')    
    
    # Save the results to dictionaries
    models[name] = clf    
    confusion[name] = confusion_matrix(y_train, y_pred)
    class_report[name] = classification_report(y_train, y_pred)
    
    results.append([name, accuracy, precision, recall, f1, f1_test, run_time.best])

    
results = pd.DataFrame(results, columns=['Classifier', 'Accuracy', 'Precision', 'Recall', 'F1 Train', 'F1 Test', 'Train Time'])
results.set_index('Classifier', inplace= True)

print ('... All done!')
```
